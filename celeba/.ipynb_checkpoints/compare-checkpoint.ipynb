```python
import numpy as np
import argparse
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from tqdm import tqdm

class LabelSmoothingCrossEntropy(nn.Module):
    def __init__(self):
        super(LabelSmoothingCrossEntropy, self).__init__()
    def forward(self, x, target, smoothing=0.3):
        smoothing = args.smoothing
        confidence = 1. - smoothing
# #         print(confidence)
# #         exit(0)
#         logprobs = torch.log_softmax(x, dim=-1)
# #         print(logprobs.shape)
# #         print(target.shape)
#         target = target.type(torch.int64)
#         nll_loss = -logprobs.gather(dim=-1, index=target)
# #         print(nll_loss)
#         nll_loss = nll_loss.squeeze(1)
#         smooth_loss = -logprobs.mean(dim=-1)
#         loss = confidence * nll_loss + smoothing * smooth_loss
#         return loss.mean()
    
        """
        写法1
        """
#         x = torch.cat((1-x, x), 1)
#         target = target.type(torch.long)
#         logprobs = F.log_softmax(x, dim=-1)
#         nll_loss = -logprobs.gather(dim=-1, index=target)
#         nll_loss = nll_loss.squeeze(1)  # 得到交叉熵损失
#         # 注意这里要结合公式来理解，同时留意预测正确的那个类，也有a/K，其中a为平滑因子，K为类别数
#         smooth_loss = -logprobs.mean(dim=1)
#         loss = confidence * nll_loss + smoothing * smooth_loss
        """
        写法2
        """
        x = torch.cat((1-x, x), 1)
#         y_hat = torch.softmax(x, dim=1)
        y_hat = x
#         print(y_hat.shape, target.shape)
        target = target.type(torch.long)
        # 这里cross_loss和nll_loss等价
        cross_loss = self.cross_entropy(y_hat, target)
#         print("???:", cross_loss.shape)
        smooth_loss = -torch.log(y_hat).mean(dim=1)
#         print(smooth_loss)
        # smooth_loss也可以用下面的方法计算,注意loga + logb = log(ab)
        # smooth_loss = -torch.log(torch.prod(y_hat, dim=1)) / y_hat.shape[1]
        loss = confidence * cross_loss + smoothing * smooth_loss
        return loss.mean()

    def cross_entropy(self, y_hat, y):
        print(y_hat[range(len(y_hat)), y].shape)
        return - torch.log(y_hat[range(len(y_hat)), y])
#         return nn.BCELoss()(y_hat, y)

l = LabelSmoothingCrossEntropy()
x = torch.tensor([[0.2, 0.8], [0.3, 0.7]])
target = [0, 1]
a = l.forward(x, target)                  
```
